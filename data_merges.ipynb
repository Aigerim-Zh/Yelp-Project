```python
import pandas as pd
import numpy as np
```

```python
business_df = pd.read_csv('CSV_files/Final_Clean_Business_Data.csv', index_col=0)
```

```python
business_df['Postal_Code'] = business_df['Postal_Code'].fillna(0) 
business_df['Postal_Code'] = business_df['Postal_Code'].astype("int64") 
business_df.drop(business_df.loc[business_df['Postal_Code']==0].index, inplace=True)
```

```python
census_df = pd.read_csv('CSV_files/econ_status_data.csv', low_memory=False)
```

```python
census_df.head()
```

```python
census_df.count()
```

```python
census_df.dtypes
```

```python
merged_df = pd.merge( business_df, census_df, left_on='Postal_Code', right_on='Postal_Code', how='left')
```

```python
merged_df.count()
```

```python
merged_df.head()
```

```python
merged_df.tail()
```

```python
merge_clean = merged_df.dropna(subset=['Median_Income(dollars)'])
```

```python
merge_clean.count()
```

```python
population_df = pd.read_csv('CSV_files/population_data.csv', low_memory=False)
```

```python
population_df.head()
```

```python
population_df.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
population_df.head()
```

```python
population_df.info()
```

```python
New_merge_df = pd.merge(merge_clean, population_df, left_on='Postal_Code', right_on='Postal_Code', how='left')
```

```python
New_merge_df.count()
```

```python
Final_merge_df = New_merge_df.dropna(subset=['Population'])
```

```python
# Dependencies
import json
import os
import pandas as pd
import numpy as np
```

```python
# read the json
business_json_path = 'Resources/yelp_academic_dataset_business.json'
business_df = pd.read_json(business_json_path, lines=True)
```

```python
#Load DataFrame
business_df
```

```python
# Columns
business_df.columns
```

```python
# Look at the categories columns
business_df.loc[:,"categories"]
```

```python
# Filter categories that contain "Restaurants"
restaurants_df = business_df[business_df['categories'].str.contains('Restaurants', case=False, na=False)]

restaurants_df
```

```python
#filter for only US data
exclude = ("AB", "XMS")
us_rests_df = restaurants_df[~(restaurants_df["state"].isin(exclude))]
us_rests_df
```

```python
#Extract the attributes to it's own column.
attributes_df = pd.concat([us_rests_df, us_rests_df["attributes"].apply(pd.Series)], axis=1)
attributes_df
```

```python
# create new dataframe with only columns relevant for our analysis 
attributes_df = attributes_df[['business_id', 'name', 'address', 'city', 'state', 'postal_code', 'latitude', 'longitude', 'stars', 'review_count', 'RestaurantsDelivery',
                               'OutdoorSeating', 'RestaurantsTakeOut', 'WiFi', 'RestaurantsReservations', 'RestaurantsGoodForGroups', 'WheelchairAccessible', 'HappyHour', 'DietaryRestrictions']]
```

```python
attributes_df
```

```python
attributes_df.dtypes
```

```python
attributes_df['DietaryRestrictions'] = attributes_df['DietaryRestrictions'].replace(np.nan, 0)
def changeStatus(status):
    if status == 0:
        return 0
    else:
        return 1
attributes_df['DietaryRestrictions'] = attributes_df['DietaryRestrictions'].apply(changeStatus)
attributes_df
```

```python
attributes_df['WiFi'] = attributes_df['WiFi'].replace(np.nan, 0)
def changeStatus(status):
    if status == 0:
        return 0
    elif status == "u'no'":
        return 0
    else:
        return 1
attributes_df['WiFi'] = attributes_df['WiFi'].apply(changeStatus)
attributes_df
```

```python
attributes_df=attributes_df.fillna(0)
attributes_df
```

```python
# Replacing 'RestaurantsDelivery' True and false with 1 and 0
def changeStatus(status):
    if status == 0:
        return 0
    elif status == 'False':
        return 0
    else:
        return 1
attributes_df['RestaurantsDelivery'] = attributes_df['RestaurantsDelivery'].apply(changeStatus)
attributes_df
```

```python
# Dependencies
import json
import os
import pandas as pd
import numpy as np
```

```python
# read the json
business_json_path = 'Resources/yelp_academic_dataset_business.json'
business_df = pd.read_json(business_json_path, lines=True)
```

```python
#Load DataFrame
business_df
```

```python
# Columns
business_df.columns
```

```python
# Look at the categories columns
business_df.loc[:,"categories"]
```

```python
# Filter categories that contain "Restaurants"
restaurants_df = business_df[business_df['categories'].str.contains('Restaurants', case=False, na=False)]

restaurants_df
```

```python
#filter for only US data
exclude = ("AB", "XMS")
us_rests_df = restaurants_df[~(restaurants_df["state"].isin(exclude))]
us_rests_df
```

```python
#Extract the attributes to it's own column.
attributes_df = pd.concat([us_rests_df, us_rests_df["attributes"].apply(pd.Series)], axis=1)
attributes_df
```

```python
# create new dataframe with only columns relevant for our analysis 
attributes_df = attributes_df[['business_id', 'name', 'address', 'city', 'state', 'postal_code', 'latitude', 'longitude', 'stars', 'review_count', 'RestaurantsDelivery',
                               'OutdoorSeating', 'BusinessAcceptsCreditCards', 'RestaurantsPriceRange2', 'Alcohol', 'GoodForKids', 'RestaurantsReservations', 'RestaurantsTakeOut', 'WiFi', 'RestaurantsGoodForGroups', 'WheelchairAccessible', 'HappyHour', 'NoiseLevel', 'DietaryRestrictions']]
attributes_df
```

```python
attributes_df=attributes_df.fillna(0)
attributes_df
```

```python
# Replacing 'RestaurantsDelivery' True and false with 1 and 0
def changeStatus(status):
    if status == 0:
        return 0
    elif status == 'False':
        return 0
    else:
        return 1
attributes_df['RestaurantsDelivery'] = attributes_df['RestaurantsDelivery'].apply(changeStatus)
attributes_df
```

```python
# Replacing 'OutdoorSeating' True and false with 1 and 0
def changeStatus(status):
    if status == 0:
        return 0
    elif status == 'False':
        return 0
    else:
        return 1
attributes_df['OutdoorSeating'] = attributes_df['OutdoorSeating'].apply(changeStatus)
attributes_df
```

```python
# Replacing 'BusinessAcceptsCreditCards' True and false with 1 and 0
def changeStatus(status):
    if status == 0:
        return 0
    elif status == 'False':
        return 0
    else:
        return 1
attributes_df['BusinessAcceptsCreditCards'] = attributes_df['BusinessAcceptsCreditCards'].apply(changeStatus)
attributes_df
```

```python
# Since price can't be 0 and None, so replace it with a 1
attributes_df['RestaurantsPriceRange2'] = attributes_df['RestaurantsPriceRange2'].replace(np.nan, 0)
def changeStatus(status):
    if status == 0:
        return 1
    elif status == "1":
        return 1
    elif status ==  "2":
        return 2
    elif status == "3":
        return 3
    elif status == "4":
        return 4
    else: 
        return 1
attributes_df['RestaurantsPriceRange2'] = attributes_df['RestaurantsPriceRange2'].apply(changeStatus)
attributes_df
```

```python
# Replacing 'Alcohol' True and false with 1 and 0, doesn't matter the type of alcohol
attributes_df['Alcohol'] = attributes_df['Alcohol'].replace(np.nan, 0)
def changeStatus(status):
    if status == 0:
        return 0
    elif status == "u'none''":
        return 0
    else:
        return 1
attributes_df['Alcohol'] = attributes_df['Alcohol'].apply(changeStatus)
attributes_df
```

```python
# Replacing 'GoodForKids' True and false with 1 and 0
def changeStatus(status):
    if status == 0:
        return 0
    elif status == 'False':
        return 0
    else:
        return 1
attributes_df['GoodForKids'] = attributes_df['GoodForKids'].apply(changeStatus)
attributes_df
```

```python
# Replacing 'RestaurantsReservations' True and false with 1 and 0
def changeStatus(status):
    if status == 0:
        return 0
    elif status == 'False':
        return 0
    else:
        return 1
attributes_df['RestaurantsReservations'] = attributes_df['RestaurantsReservations'].apply(changeStatus)
attributes_df
```

```python
# Replacing 'RestaurantsTakeOut' True and false with 1 and 0
def changeStatus(status):
    if status == 0:
        return 0
    elif status == 'False':
        return 0
    else:
        return 1
attributes_df['RestaurantsTakeOut'] = attributes_df['RestaurantsTakeOut'].apply(changeStatus)
attributes_df
```

```python
attributes_df['WiFi'] = attributes_df['WiFi'].replace(np.nan, 0)
def changeStatus(status):
    if status == 0:
        return 0
    elif status == "u'no'":
        return 0
    else:
        return 1
attributes_df['WiFi'] = attributes_df['WiFi'].apply(changeStatus)
attributes_df
```

```python
# Replacing 'RestaurantsGoodForGroups' True and false with 1 and 0
def changeStatus(status):
    if status == 0:
        return 0
    elif status == 'False':
        return 0
    else:
        return 1
attributes_df['RestaurantsGoodForGroups'] = attributes_df['RestaurantsGoodForGroups'].apply(changeStatus)
attributes_df
```

```python
# Replacing 'WheelchairAccessible' True and false with 1 and 0
def changeStatus(status):
    if status == 0:
        return 0
    elif status == 'False':
        return 0
    else:
        return 1
attributes_df['WheelchairAccessible'] = attributes_df['WheelchairAccessible'].apply(changeStatus)
attributes_df
```

```python
# Replacing 'HappyHour' True and false with 1 and 0
def changeStatus(status):
    if status == 0:
        return 0
    elif status == 'False':
        return 0
    else:
        return 1
attributes_df['HappyHour'] = attributes_df['HappyHour'].apply(changeStatus)
attributes_df
```

```python
attributes_df['NoiseLevel'] = attributes_df['NoiseLevel'].replace(np.nan, 0)
def changeStatus(status):
    if status == 0:
        return 1
    elif status == "u'quiet'":
        return 1
    elif status ==  "'quiet'":
        return 1
    elif status == "u'average''":
        return 2
    elif status == "'average'":
        return 2
    elif status == "u'loud'":
        return 3
    elif status == "'loud'":
        return 3
    else:
        return 4
attributes_df['NoiseLevel'] = attributes_df['NoiseLevel'].apply(changeStatus)
attributes_df
```

```python
attributes_df['DietaryRestrictions'] = attributes_df['DietaryRestrictions'].replace(np.nan, 0)
def changeStatus(status):
    if status == 0:
        return 0
    else:
        return 1
attributes_df['DietaryRestrictions'] = attributes_df['DietaryRestrictions'].apply(changeStatus)
attributes_df
```

```python
# Using .rename(columns={}) in order to rename columnsattributes_df
renamed_df = attributes_df.rename(columns={"business_id":"Restaurant_ID", "name":"Restaurants_Name", "address":"Address","city":"City",
                                          "state":"State","postal_code":"Postal_Code","latitude":"Latitude","longitude":"Longitude","stars":"Stars_Rating",
                                           "review_count":"Review_Count","RestaurantsDelivery":"Restaurants_Delivery","OutdoorSeating":"Outdoor_Seating", "BusinessAcceptsCreditCards":"Accepts_CreditCards", "RestaurantsPriceRange2":"Price_Range", "Alcohol":"Alcohol", "GoodForKids":"Good_For_Kids", "RestaurantsReservations":"Reservations", "RestaurantsTakeOut":"Restaurants_TakeOut",
                                           "WiFi":"WiFi", "RestaurantsGoodForGroups":"Good_For_Groups", "WheelchairAccessible":"Wheelchair_Accessible", "HappyHour":"Happy_Hour", "NoiseLevel":"Noise_Level", "DietaryRestrictions":"Dietary_Restrictions"})
renamed_df.head()
```
